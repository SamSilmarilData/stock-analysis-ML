# Visualisation

### the closing prices over time

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,7))
adbe_data['Close'].plot()
plt.title('Adobe Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

### volume of stocks traded over time.

```python
plt.figure(figsize=(14,7))
adbe_data['Volume'].plot(color='gray')
plt.title('Adobe Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

## Moving averages

### Calculate and plot short-term and long-term moving averages to understand trends.

```python
adbe_data['50_MA'] = adbe_data['Close'].rolling(window=50).mean()
adbe_data['200_MA'] = adbe_data['Close'].rolling(window=200).mean()

plt.figure(figsize=(14,7))
adbe_data[['Close', '50_MA', '200_MA']].plot(ax=plt.gca())
plt.title('Adobe Closing Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

## Outliers

```python
plt.figure(figsize=(8,6))
adbe_data['Close'].plot(kind='box')
plt.title('Boxplot of Adobe Closing Prices')
plt.ylabel('Price')
```

## Correlation Analysis

```python
correlation_matrix = adbe_data.corr()
import seaborn as sns
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Adobe Stock Data')
```

## Distribution of Data

```python
plt.figure(figsize=(10,6))
sns.histplot(adbe_data['Close'], kde=True)
plt.title('Distribution of Adobe Closing Prices')
plt.xlabel('Closing Price')
```
